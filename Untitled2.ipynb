```python
def quantile_in_stock(ss):
    return ss.groupby('date').rank(pct=True)

def assess_var(df,var,ycol):
    df2 = df[[var,ycol]].copy()
    var_q = '{}_q'.format(var)
    print(var_q)
    df2[var_q] = quantile_in_stock(df[var])
    qbins = 20
    df2['bin'] = pd.qcut(df2[var_q],qbins)
    
    tmp = df2.groupby(['date','bin'])[ycol].mean()

    tmp = df2.groupby(['date','bin'])[ycol].mean()
    tmp.shape
    tmp.head()
    tmp_tot = tmp.groupby('bin').agg(['mean','median','max','min','std'])
    
    tmp_mth = tmp.reset_index('bin').groupby('bin').resample('Q').agg('mean').pivot_table(index='bin',columns='date',values=ycol)
    
    return tmp_tot,tmp_mth


def stats(df,final_mask,date_index,holding_length=[1,2,3,5,10,20,60]):
    r_cols = ['r{}'.format(i) for i in holding_length]
    print(df.loc[final_mask,r_cols].groupby('date').agg(['mean','median','std']).shape)
    tmp_day = df.loc[final_mask,r_cols].groupby('date').agg(['mean','median','std'])
    
    fee_rate = 0.0014
#     tmp_day = (tmp_day-fee_rate).reindex(date_index)
    tmp_day = (tmp_day-fee_rate).reindex(date_index).fillna(0)
#     for i in [2,5,10,20,60]:
# #         tmp_day[('r{}'.format(i),'mean')]=tmp_day[('r{}'.format(i),'mean')].fillna(0)
#         display(tmp_day[[('r{}'.format(i),'mean')]].agg(['mean','median','std','max','min']))

#     print(tmp_day.head())
#     print(tmp_day.shape,tmp_day['r5'].notnull().sum(),'---------')
#     print((tmp_day+1).resample('M').last().cumprod().dropna().iloc[-1])
    print(min(tmp_day.index),max(tmp_day.index))
    # tmp_day.tail(100)
    
    res_list = []
    plt.figure()
    for k in holding_length:
        r = []
        for i in range(k):
            tmp_r = (tmp_day[('r{}'.format(k),'mean')].iloc[range(i,len(tmp_day),k)]+1)
    #         tmp_r.agg('mean'),tmp_r.cumprod().dropna().iloc[-1]
#             r.append(tmp_r.cumprod().dropna().iloc[-1])
            tmp_r.name = 'r{}_{}'.format(k,i)
#             tmp_r.head()
            r.append(tmp_r)
        df_r = pd.concat(r,axis=1,join='outer').fillna(1).shift(k)
#         df_r.columns = ['r{}_{}'.format(k,i) for i]
        df_r.shape
        print(df_r.tail())
#         df_r_avg = df_r.sum(axis=1)
        print(df_r.resample('A').prod())
        final_r = df_r.prod()
        print('\n',k,final_r.sum()/k, np.std(final_r))
        
        df_nv = df_r.cumprod()
        df_nv['r{}'.format(k)] = df_nv.mean(axis=1)
        res_list.append(pd.concat([df_r,df_nv.add_prefix('nv_')],axis=1,join='inner'))
        plt.plot(df_nv.index,df_nv['r{}'.format(k)].values)
    
    return pd.concat(res_list,axis=1)
```
